```python
# Define a list of features that are related to infertility causes
important_feat = [
    'V2',
    'V4',
    'V8',
    'V11',
    'Amount'
]
```

```python
# Split the data into training and testing sets
X = df[important_feat]
y = df['Class']
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import cross_val_predict, StratifiedKFold
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from imblearn.over_sampling import RandomOverSampler, SMOTE
from imblearn.under_sampling import RandomUnderSampler

# Define imbalance handling techniques
techniques = [
    ("Random Oversampling", RandomOverSampler(sampling_strategy='auto', random_state=42)),
    ("Random Undersampling", RandomUnderSampler(sampling_strategy='auto', random_state=42)),
    ("SMOTE", SMOTE(sampling_strategy='auto', random_state=42))
]

# Define models
models = [
    ("Logistic Regression", LogisticRegression(random_state=42)),
    ("Random Forest", RandomForestClassifier(n_estimators=100, random_state=42)),
    ("Decision Tree", DecisionTreeClassifier(random_state=42))
]

# Define evaluation metrics
evaluation_metrics = {
    "Accuracy": accuracy_score,
    "Precision": precision_score,
    "Recall": recall_score,
    "F1 Score": f1_score
}

# Initialize a list to store results DataFrames
results_dfs = []

# K-fold Cross-Validation
num_folds = 5
kfold = StratifiedKFold(n_splits=num_folds, shuffle=True, random_state=42)

for technique_name, technique in techniques:
    for model_name, model in models:
        # Initialize lists to store evaluation metric scores for each fold
        metric_scores = []
        
        for train_index, test_index in kfold.split(X, y):
            X_train_fold, X_test_fold = X.iloc[train_index], X.iloc[test_index]
            Y_train_fold, Y_test_fold = y.iloc[train_index], y.iloc[test_index]
            
            # Apply the imbalance handling technique
            X_train_resampled, Y_train_resampled = technique.fit_resample(X_train_fold, Y_train_fold)
            
            # Create and train the model
            model.fit(X_train_resampled, Y_train_resampled)
            
            # Make predictions for the current fold
            Y_pred_fold = model.predict(X_test_fold)
            
            # Calculate evaluation metrics for the current fold
            metric_scores.append([eval_metric(Y_test_fold, Y_pred_fold) for eval_metric in evaluation_metrics.values()])
        
        # Compute the mean of the metric scores across folds
        mean_metric_scores = np.mean(metric_scores, axis=0)
        
        # Create a DataFrame for the results of this technique and model
        results_df = pd.DataFrame({
            "Technique": [technique_name],
            "Model": [model_name],
            **{metric_name: [score] for metric_name, score in zip(evaluation_metrics.keys(), mean_metric_scores)}
        })
        
        # Append the results DataFrame to the list
        results_dfs.append(results_df)

# Concatenate all results DataFrames into one
results_table = pd.concat(results_dfs, ignore_index=True)

# Print the results table
results_table
```
